```python
import requests
import pandas as pd
import io
from datetime import date
```

```python
CEA_dataset_id = "d_ee7e46d3c57f7865790704632b0aef71"

response = requests.get(f'https://api-open.data.gov.sg/v1/public/api/datasets/{CEA_dataset_id}/initiate-download')
```

```python
data = requests.get(response.json()['data']['url'])
```

```python
df = pd.read_csv(io.BytesIO(data.content))
df["transaction_date"] = (
    pd.to_datetime(df["transaction_date"], format="%b-%Y")
      .dt.date
)
```
